# Run tests on data file


```python
import json
with open('data.json', 'r') as f:
    data = json.load(f)
```

```python
def check_correct_format(text):
    parts = text.split("\n")
    flag = True
    for part in parts:
        if not part:
            continue
        if not part.startswith("<|system|>") and not part.startswith("<|user|>") and not part.startswith("<|assistant|>"):
            if not part.startswith("<|system|>"):
                flag = False
                print("Error: System utterance must start with '<|system|>'. Fixed automatically.")
            if not part.startswith("<|user|>"):
                flag = False
                print("Error: User utterance must start with '<|user|>'. Fixed automatically.")
            if not part.startswith("<|assistant|>"):
                flag = False
                print("Error: Assistant utterance must start with '<|assistant|>'. Fixed automatically.")
        if not part.endswith("</s>"):
            flag = False
            print("Error: Utterance must end with '</s>'. Fixed automatically.")
        
    return flag
```

```python
def correct_format(text):
    parts = text.split("\n")
    for i in range(len(parts)):
        part = parts[i]
        if not part:
            continue
        if not part.startswith("<|system|>") and not part.startswith("<|user|>") and not part.startswith("<|assistant|>"):
            if not part.startswith("<|system|>"):
                part = "<|system|>" + part
                print("Error: System utterance must start with '<|system|>'. Fixed automatically.")
            if not part.startswith("<|user|>"):
                part = "<|user|>" + part
                print("Error: User utterance must start with '<|user|>'. Fixed automatically.")
            if not part.startswith("<|assistant|>"):
                part = "<|assistant|>" + part
                print("Error: Assistant utterance must start with '<|assistant|>'. Fixed automatically.")
        if not part.endswith("</s>"):
            part += "</s>"
            print("Error: Utterance must end with '</s>'. Fixed automatically.")
        parts[i] = part
    return "\n".join(parts)
```

## Run tests on data format


```python
wrong_format = []
```

```python
for d in data:
    new_text = correct_format(d['response']['text'])
    d['response']['text'] = new_text
```

## Fix data format


```python
for d in data:
    if not check_correct_format(d['response']['text']):
        wrong_format.append(d['data_id'])
```
